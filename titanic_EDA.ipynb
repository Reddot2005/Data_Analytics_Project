```python
#importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
titanic=pd.read_csv('train.csv',index_col="PassengerId")
```

```python
#view
titanic
titanic.shape
titanic.apply(pd.isnull).sum()
```

```python
#coverting categorical to numerical using mapp function
titanic['Sex']=titanic['Sex'].map({'male':0,'female':1})
```

```python
#view number of columns
titanic.columns
```

```python
#checking the datatype of sex
titanic["Sex"].dtypes
```

```python
#view number of rows and columns
titanic.shape
```

```python
titanic["Age"]>35
```

```python
#droping cabin column
titanic=titanic.drop("Cabin",axis=1)
titanic.shape
```

```python
titanic
```

```python
#cheching the number of null values in each column
titanic.apply(pd.isnull).sum()
```

```python
#filling missing age values with the median
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
```

```python
titanic
```

```python
titanic.apply(pd.isnull).sum()
```

```python
titanic['Embarked'].fillna(titanic['Embarked'].mode()[0], inplace=True)
```

```python
titanic.apply(pd.isnull).sum()
```

```python
# Grouping data to calculate survival counts based on gender
survived_by_gender = titanic[titanic['Survived'] == 1]['Sex'].value_counts()

# Pie chart visualization
plt.figure(figsize=(8, 8))
colors = ['skyblue', 'lightcoral']
plt.pie(survived_by_gender, labels=survived_by_gender.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=[0.1, 0], textprops={'fontsize': 12})
plt.title('Pie Chart of Survival by Gender', fontsize=14)
plt.axis('equal')  # Equal aspect ratio ensures the pie is drawn as a circle.

# Display the plot
plt.show()
```

```python
# Create scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(titanic['Sex'], titanic['Survived'], c='r', cmap='viridis', edgecolors='black')
plt.title('Sex vs Survived')
plt.xlabel('Sex (0 = Male, 1 = Female)')
plt.ylabel('Survived (0 = Dead, 1 = Survived)')
plt.yticks([0, 1], ['Dead', 'Survived'])
plt.xticks([0, 1], ['Male', 'Female'])
plt.grid(True)
plt.show()
```

```python


# Create cross-tabulation of Sex and Survival
cross_tab = pd.crosstab(titanic['Sex'], titanic['Survived'])

# Create heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cross_tab, annot=True, fmt='d')

plt.show()
```

```python
# Survival by gender
plt.subplot(2, 2, 1)
sns.countplot(x='Sex', hue='Survived', data=titanic)
plt.title('Survival by Gender')
```

```python
# Age distribution and survival
plt.subplot(2, 2, 3)
sns.histplot(titanic, x='Age', hue='Survived', kde=True, bins=30)
plt.title('Age Distribution and Survival')
```

```python
# Survival by class
plt.subplot(2, 2, 2)
sns.countplot(x='Pclass', hue='Survived', data=titanic)
plt.title('Survival by Passenger Class')
```

```python
# Fare distribution and survival
plt.subplot(2, 2, 4)
sns.histplot(titanic, x='Fare', hue='Survived', kde=True, bins=30)
plt.title('Fare Distribution and Survival')

plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap
plt.figure(figsize=(10, 6))
correlation = titanic[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.rugplot(data=titanic, x='Survived')
plt.show()

# More customized rugplot
plt.figure(figsize=(10, 4))
sns.rugplot(data=titanic, x='Survived',
            height=0.5,          # Height of the lines
            color='red',         # Color of the lines
            alpha=0.5)           # Transparency of the lines
plt.title('Distribution of Values')
plt.xlabel('Values')
plt.show()
```

```python

# FacetGrid (FacetGrid allows flexible multi-plot grids)
g = sns.FacetGrid(titanic, col="Survived", row="Sex", margin_titles=True)
g.map(sns.histplot, "Age", kde=True, bins=20)
g.fig.suptitle("FacetGrid: Age Distribution by Survival and Gender", y=1.02)
plt.show()
```

```python
# Jointplot
sns.jointplot(x='Age', y='Survived', data=titanic, kind='scatter', hue='Sex')
plt.suptitle("Jointplot: Age vs Survival", y=1.02)
plt.show()
```

```python
# Create the box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='Survived', y='Age', data=titanic)

# Add labels and title
plt.title('Box Plot of Age by Survival Status', fontsize=14)
plt.xlabel('Survived (0 = No, 1 = Yes)', fontsize=12)
plt.ylabel('Age', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Example: Count plot for Survived vs Gender
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='Survived', hue='Sex', palette='Set2')

# Adding labels and title
plt.title('Countplot of Survival by Gender', fontsize=14)
plt.xlabel('Survived', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Gender')

# Display the plot
plt.show()
```

```python
# Plotting the histogram with a polygon
plt.figure(figsize=(8, 6))
sns.histplot(data=titanic, x='Age', hue='Survived', kde=True, element='step', stat='density', common_norm=False)

# Adding labels and title
plt.title('Histogram Polygon of Age by Survival', fontsize=14)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Survived', labels=['Did Not Survive (0)', 'Survived (1)'])

# Display the plot
plt.show()
```

#conclusions and insights

Key Findings:

Gender (Sex) as Primary Survival Predictor:


Being female was the strongest predictor of survival on the Titanic
Women had significantly higher survival rates compared to men
This aligns with the historical "women and children first" protocol followed during the disaster


Age as Secondary Factor:


Age played a notable but lesser role in survival outcomes
Young children showed better survival rates
Middle-aged men had the lowest survival probability
The age distribution of survivors skewed younger than that of non-survivors


Limited Impact of Other Variables:


Other parameters like passenger class, cabin location, and other demographic factors showed less significant correlations with survival
While these variables might have had some influence, their predictive power was substantially lower than sex and age
